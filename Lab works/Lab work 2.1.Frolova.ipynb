#### Завдання 5

Напишіть програму для обчислення коренів квадратного рівняння.

Нагадую, що `n`-го степеня має `n` коренів. (Корені можуть бути комплексні)

```python
import cmath
import math
a = int(input())
b = int(input())
c = int(input())
if a == 0 and b == 0 and c == 0:
    print('x может принимать любые значения')
elif a == 0 and b == 0:
    print('Корней нет')
else:
    if a == 0 and c == 0:
        print('x =', 0)
    else:
        if a == 0:
            x = (- c / b)
            print('x =', x)
        elif b == 0:
            x = math.sqrt(- c / a)
            print('x =', x)
        else:
            discr = b * b - 4 * a * c
            if discr == 0:
                x = - b / (2 * a)
                print('x =', x)
            else:
                x1 = (-b + cmath.sqrt(discr)) / (2 * a)
                x2 = (-b - cmath.sqrt(discr)) / (2 * a)
                print('x1 =', x1, 'x2 =', x2)
```
