## Імпортування

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
# Цей рядок призначений лише для блокнотів jupyter.
# Якщо ви використовуєте інший редактор, то використовуйте: plt.show () 
# в кінці всіх ваших команд побудови графіка, щоб фігура появлялась в іншому вікні.
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

#### Основні команди Matplotlib

```python
plt.plot(x, y, 'b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('First')
```

#### Декілька кординатних площин на одному полотні

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'g--') 
plt.subplot(1,2,2)
plt.plot(y, x, 'r-')
```

#### Обєктно-орієнтовний підхід

Об'єкт plt.subplots () працює як автоматичний менеджер осей.


```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
plt.tight_layout()
```

```python
fig, ax = plt.subplots(2)


ax[0].plot(x, x**2)


ax[1].plot(x, x**(1/2));
```

#### Легенда, заголовок, мітки даних

```python
fig, ax = plt.subplots()

ax.set_title("Заголовок");
ax.set_xlabel("x")
ax.set_ylabel("y");
ax.plot(x, x**2, label="x**2")
ax.plot(x, x**3, label="x**3")
ax.legend(loc=1) #параметр loc визначає розміщення легенди
```

#### Розмір полотна

```python
# fig = plt.figure(figsize=(8,1), dpi=100)
fig, axes = plt.subplots(figsize=(12,4))

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Заголовок');
```

#### Відображення ліній осей

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))


axes[0].plot(x, x**2, x, x**3, lw=1)
axes[0].grid(True)


axes[1].plot(x, x**2, x, x**3)
axes[1].grid(color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)
```
